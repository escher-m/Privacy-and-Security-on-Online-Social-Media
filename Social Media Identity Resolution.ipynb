```python
import pandas as pd
```

```python
df=pd.read_csv('/Users/karishma/Downloads/datasets/IdentityResolution.csv')
df.head(10)
```

```python
df.columns=['Name','Twitter','Facebook','Instagram']
```

```python
df.head(10)
```

```python
df1=df
```

```python
def find_username_Twitter(row):
    link=str(row['Twitter'])
    return link[link.rindex('/')+1:]
```

```python
def find_username_Facebook(row):
    link=str(row['Facebook'])
    return link[link.rindex('/')+1:]
```

```python
def find_username_Instagram(row):
    link=str(row['Instagram'])
    return link[link.rindex('/')+1:]
```

```python
df1['Twitter_Username'] = df1.apply (lambda row: find_username_Twitter(row), axis=1)
df1['Facebook_Username'] = df1.apply (lambda row: find_username_Facebook(row), axis=1)
df1['Instagram_Username'] = df1.apply (lambda row: find_username_Instagram(row), axis=1)

df1
```

```python
df_username = df1[['Name', 'Twitter_Username', 'Facebook_Username','Instagram_Username']].copy()
```

```python
df_username
```

```python
import Levenshtein as lv
import enchant
```

```python
def calc_lv_TF(row):
    lv_TF=lv.distance(str(row['Twitter_Username']),str(row['Facebook_Username']))
    return lv_TF
```

```python
#df_username['lv_TF'] = df_username.apply (lambda row: calc_lv_TF(row))
```

```python
def calc_lve_TF(row):
    lve_TF=enchant.utils.levenshtein(str(row['Twitter_Username']),str(row['Facebook_Username']))
    return lve_TF
```

```python
def calc_lve_FI(row):
    lve_FI=enchant.utils.levenshtein(str(row['Facebook_Username']),str(row['Instagram_Username']))
    return lve_FI
```

```python
def calc_lve_TI(row):
    lve_TI=enchant.utils.levenshtein(str(row['Twitter_Username']),str(row['Instagram_Username']))
    return lve_TI
```

```python
df_username['lve_TF'] = df_username.apply (lambda row: calc_lve_TF(row), axis=1)
```

```python
df_username['lve_FI'] = df_username.apply (lambda row: calc_lve_FI(row), axis=1)
```

```python
df_username['lve_TI'] = df_username.apply (lambda row: calc_lve_TI(row), axis=1)
```

```python
df_username
```

```python
from pyjarowinkler import distance as jw
```

```python
def calc_jw_TF(row):
    jw_TF=jw.get_jaro_distance(row['Twitter_Username'],row['Facebook_Username'],winkler=True,scaling=0.1)

    return jw_TF
```

```python
def calc_jw_FI(row):
    jw_FI=jw.get_jaro_distance(row['Facebook_Username'],row['Instagram_Username'],winkler=True,scaling=0.1)

    return jw_FI
```

```python
def calc_jw_TI(row):
    jw_TI=jw.get_jaro_distance(row['Twitter_Username'],row['Instagram_Username'],winkler=True,scaling=0.1)

    return jw_TI
```

```python
df_username['jw_TF'] = df_username.apply (lambda row: 999 if not str(row['Twitter_Username']) or not str(row['Facebook_Username']) else calc_jw_TF(row), axis=1)
```

```python
df_username['jw_FI'] = df_username.apply (lambda row: 999 if not str(row['Facebook_Username']) or not str(row['Instagram_Username']) else calc_jw_FI(row), axis=1)
```

```python
df_username['jw_TI'] = df_username.apply (lambda row: 999 if not str(row['Twitter_Username']) or not str(row['Instagram_Username']) else calc_jw_TI(row), axis=1)
```

```python
df_username
```

```python
mean_lve_TF= df_username["lve_TF"].mean()
std_lve_TF=df_username["lve_TF"].std()

mean_lve_FI= df_username["lve_FI"].mean()
std_lve_FI=df_username["lve_FI"].std()

mean_lve_TI= df_username["lve_TI"].mean()
std_lve_TI=df_username["lve_TI"].std()

mean_jw_TF= df_username["jw_TF"].mean()
std_jw_TF=df_username["jw_TF"].std()

mean_jw_FI= df_username["jw_FI"].mean()
std_jw_FI=df_username["jw_FI"].std()

mean_jw_TI= df_username["jw_TI"].mean()
std_jw_TI=df_username["jw_TI"].std()
```

```python
data_lve = {'Mean':[mean_lve_TF,mean_lve_FI,mean_lve_TI], 
        'StandardDeviation':[std_lve_TF,std_lve_FI,std_lve_TI]} 
data_lve_plot = pd.DataFrame(data_lve, index =['Twitter-Facebook','Facebook-Instagram',
                                'Twitter-Instagram']) 
data_lve_plot
```

```python
data_jw = {'Mean':[mean_jw_TF,mean_jw_FI,mean_jw_TI], 
        'StandardDeviation':[std_jw_TF,std_jw_FI,std_jw_TI]} 
data_jw_plot = pd.DataFrame(data_jw, index =['Twitter-Facebook','Facebook-Instagram',
                                'Twitter-Instagram']) 
data_jw_plot
```

```python
lvplot=data_lve_plot.plot(kind = "bar", y = "Mean", legend = False, 
            title = "Mean and Standard Deviation of Levenshtein Distance Metric", yerr = "StandardDeviation", figsize=(10,6),fontsize=14,ylabel='Similarity of Usernames')
lvplot.yaxis.label.set_color('blue')
lvplot.xaxis.label.set_color('blue')

lvplot.set_xlabel("Social Media", size=14)
lvplot.set_ylabel("Similarity of Usernames ",size=14)

```

```python
jwplot=data_jw_plot.plot(kind = "bar", y = "Mean", legend = False, 
            title = "Mean and Standard Deviation of Jaro Winkler Distance Metric", yerr = "StandardDeviation", figsize=(10,6),fontsize=14,ylabel='Similarity of Usernames')
jwplot.yaxis.label.set_color('blue')
jwplot.xaxis.label.set_color('blue')

jwplot.set_xlabel("Social Media", size=14)
jwplot.set_ylabel("Similarity of Usernames ",size=14)
```
